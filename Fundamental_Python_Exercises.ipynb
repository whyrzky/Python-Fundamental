**CONDITIONAL STATEMENT EXERCISE**

Di mana kita memberikan suatu kondisi yang kita inginkan untuk mendapatkan suatu yang valid

- Menggunakan IF yang dapat dianalogikan dengan *"Jika sesuatu memenuhi suatu kondisi maka lakukanlah A, namun jika tidak maka tidak terjadi apa-apa"*.
- Menggunakan IF dan ELSE yang dapat direpresentasikan dengan *"Jika sesuatu memenuhi suatu kondisi maka lakukanlah A, namun jika tidak terpenuhi maka lakukanlah B"*.
- Menggunakan IF, ELIF dan ELSE. ELIF merupakan singkatan dari ELSE IF di python. Penerapan konstruksi IF, ELIF dan ELSE ini dapat dianalogikan dengan "Jika sesuatu memenuhi suatu kondisi maka lakukanlah A, jika tidak, lakukanlah pengecekan pada kondisi berikutnya yaitu jika memenuhi lakukanlah B, jika tidak maka lakukanlah C".

```python
#PRAKTEK DENGAN IF
i = 10 #inisialisasi variable i yang memiliki nilai 10

if(i==10): #pengecekan nilai i apakah sama dengan 10
    print("ini adalah angka 10") #jika TRUE maka akan mencetak kalimat ini
```

```python
angka = 5 #defiinisi variable
if angka % 2 != 0 : #perintah pertama
    print(angka,'adalah bilangan ganjil')
```

```python
#PRAKTEK DENGAN IF ELSE

angka = 4
if angka % 2 == 0:
  print(angka, "adalah bilangan genap")
else : #else hanya dieksekusi apabila statement di atas tidak terpenuhi
  print(angka,"adalah bilangan ganjil")
```

```python
i = 5
if i == 10:
	print('ini adalah angka 10')
else:
	print('bukan angka 10')
```

```python
#PRAKTEK DENGAN ELIF Contoh 1

i=50

if i==5:
	print("ini adalah angka 5")
elif i>5:
	print('lebih besar dari 5')
else:
	print('lebih kecil dari 5')
```

```python
#PRAKTEK DENGAN ELIF Contoh 2
#elif itu singkatan dari Else and If, dia akan meriksa  setiap elif (dimulai dari paling atas) sampai ketemu kondisi yang valid

rev = 1000
if rev < 300:
  print("UMKM Skala Mikro")
elif 300 <= rev < 500:
  print("UMKM Skala Kecil")
elif 500 <= rev <= 1000:
  print("UMKM Skala Menengah")
elif rev > 1000:
  print("UMKM Skala Besar")
```

```python
#PRAKTEK DENGAN ELIF Contoh 3
#Ini hanya menyederhanaan dari Contoh 1 yang setiap ELIF itu ada redundant
#Kenapa pakai ELSE di statement terakhir? karena jika semua elif tidak terpenuhi, artinya REV > 1000 (tanpa kita harus state rev harus > 1000)

rev = 1200
if rev < 300:
  print("UMKM Skala Mikro")
elif rev < 500:
  print("UMKM Skala Kecil")
elif rev <= 1000:
  print("UMKM Skala Menengah")
else:
  print("UMKM Skala Besar")
```

```python
#PRAKTEK DENGAN NESTED IF
#Fungsi IF sendiri dapat digunakan secara bertingkat atau dapat memiliki pengecekan lebih dari 1 kondisi, sebagai contoh: if ( i<7 && i <3)
#Pernyataan ini berarti bahwa i harus bernilai kurang dari 7 dan juga harus kurang dari 3 agar bisa memenuhi pengecekan tersebut.

i=2
if (i<7):
	print("nilai i kurang dari 7")
	if (i<3):
		print("nilai i kurang dari 7 dan kurang dari 3")
	else:
		print("nilai i kurang dari 7 tapi lebih besar dari 3")
```

**FUNGSI**

Membuat fungsi dengan 'def', gunanya adalah ketika kita ingin membuat output namun memerlukan beberapa rumus, maka kita cukup memanggil fungsi tersebut yang sudah kita buat di awal.

```python
#1 kita buat fungsinya dulu
def luas_persegi_panjang (panjang,lebar): #<<<<<< ini adalah FUNGSI
    hitung = panjang*lebar #<<<<<<< ini adalal LOKAL VARIABEL
    return hitung
```

**OBJECT**

adalah kumpulan kondisi dan aksi pada suatu konteks. Di mana aksi akan mempengaruhi kondisi. Pada Python:
 - Aksi = Method
 - Kondisi = Attribute

Both of them dipisahkan oleh simbol titik "."

Syntaxnya adalah:
(attributes).(method)

```python
#membuat method secara manual harus diawali dengan 'class' >> buat fungsi 'def'

class baju:
  def __init__(self,warna,model): #<<<<< perhatikan ada 'self', itu harus ada
    self.warna = warna
    self.model = model
```

```python
#untuk mengoperasikan method di atas, kita membuat list baru terlebih dahulu yaitu baju_andi dan baju_aldo

baju_andi = baju('merah','oneck')
baju_aldo = baju('hitam','vneck')
```

```python
#selanjutnya kita bisa menggunakan method dengan simbol titik "." untuk setiap list, contoh

baju_aldo.warna
```

```python
#kita juga bisa membuat lebih dari 1 method dalam 'class baju'

class baju:
  def __init__(self,warna,model):   #<<<< perhatikan ada '__init__', fungsinya utk memulai suatu method
    self.warna = warna
    self.model = model
  def kotor(self):
    print('Baju Harus Dicuci') #<<< berisi STRING

#So coba kita panggil method 'kotor'-nya untuk baju_andi
baju_andi.kotor()    #<<<<< karena berisi STRING maka harus menggunakan kurung
```

```python
#kita juga bisa membuat method dapat mengubah suatu attribute

class baju:
  def __init__(self,warna,model):
    self.warna = warna
    self.model = model
  def kotor(self):
    print('Harus Dicuci')
  def gantiwarna(self,warna_baru):
    self.warna = warna_baru #<<<<<< 'self.warna' disini dikawinkan dengan 'self.warna' pada 'def __init__' agar method 'warna_baru' dapat mengubah attribute 'warna'

#lalu kita mau ganti warna bajunya baju_andi menjadi abu tua
baju_andi.gantiwarna('abu tua')

#maka kita bisa tampilkan warna terbaru baju_andi
baju_andi.warna
```

##Deal with Missing Values

Apa itu Missing Values?
- NaN = Not a Number (baik di kolom dan row)

Kenapa bisa terjadi?
1. Structural Missing = sistem tidak mencatat data (co: Jika ada nasabah yang tidak memiliki nomor kartu kredit, maka nasabah tersebut bisa kita justifikasi tidak memiliki kartu kredit --> delete row)
2. Random Missing = bagian dari data menghilang tidak terduga (co: Human error tidak senagaja menghapus data, atau system down)

How to deal it?

##How to Overcome the Outlier

Outlier = adalah sebuah data yang memiliki karakteristik berbeda dengan data lainnya (contoh: anak 12 tahun tercatat sebagai karyawan, padahal 12 tahun adalah umur yang tidak wajar menjadi karyawan)

Sebab outlier:
1. Karena ada fenomena yang unik (ex: ada ujian yang sangat sulit di kampus, namun hanya ada 1 mahasiswa yang punya nilai 90, rata-rata mahasiswa nilai 50)
2. Karena kesalahan sistem (ex: Human error mengisi umur menjadi 99 pada data mahasiswa)

How to handle that:
1. Outlier tidak harus dibuang, karena mungkin akan terjadi di masa depan nanti
2. Outlier bisa dibuang jika hanya ada outlier extreme yang jumlahnya sedikit

###Outlier Numerical

```python
df_titanic.describe()
```

```python
df_titanic.sort_values('Fare')
```

Dalam analisis deskriptif, nilai Fare (harga tiket) tertinggi adalah 512, hal itu bukan outlier karena bukan suatu fenomena yang aneh, melainkan nyata adanya. 

Di sisi lain, ada Fare (harga tiket) yang 0 padahal Pclass berbeda-beda. Kita dapat pahami bahwa Pclass adalah peringkat/kelas, maka seharusnya penumpang yang ada pada Pclass 1 dan Fare (harga tiket) 0 adalah outlier, karena tidak masuk akal.

Untuk treatment outlier tersebut. Stepnya adalah:
1. Filter Fare > 0
2. Group by Pclass on average of each Fare (ex: Pclass 1 rerata XXX, Pclass 2 rerata XXX)
3. Concat Fare 0 tersebut dengan rerata sesuai grup Pclass

```python
#Filter Fare > 0 
filter_fare = df_titanic[df_titanic['Fare']>0]
filter_fare
```

```python
#Groupby Pclass with average of each Pclass from filter_fare
filter_fare.groupby('Pclass').agg(({'Fare':'mean'}))
```

```python
#Concat fare 0 with mean Fare based on its Pclass
df_titanic.loc[(df_titanic['Pclass']==1) & (df_titanic['Fare']==0),'Fare'] = 86.148874
df_titanic.loc[(df_titanic['Pclass']==2) & (df_titanic['Fare']==0),'Fare'] = 21.358661
df_titanic.loc[(df_titanic['Pclass']==3) & (df_titanic['Fare']==0),'Fare'] = 13.787875
df_titanic.sort_values('Fare')
```

#Batch 9: Data Encoding

Data Encoding = adalah proses konversi dari categorical data menjadi numerical data tanpa merubah esensi data kategori tersebut. Gunanya adalah agar data kategori bisa diproses dalam machine learning.

Jenis Encoding:
- On Hot Encoding = ketika kita memiliki 2 atau lebih kategori namun kategori tersebut tidak memiliki hirarki/urutan satu sama lain (ex: nama Kota, gender, dll) contoh output:

image.png
- Label Encoding = ketika kita memiliki lebih dari 2 kategori namun kategori tersebut memiliki hirarki/urutan satu sama lain (ex: kelas, pangkat, dll) contoh output:

image.png
- Target Mean Encoding = ketika kita memiliki banyak sekali kategori namun kategori tersebut memiliki hirarki/urutan satu sama lain secara Mean. Kita bisa mengkategorikannya berdasarkan rataan kejadian pada variable target terhadap masing-masing kategori tersebut pada variable input. Contoh output:

image.png


- Binary One Hot Encoding:
hanya bisa digunakan utk kategori yang bersifat biner atau 2 pilihan, contoh Gender. Maka encode menjadi 1 dan 0 adalah tidak masalah dan bukan urutan.

```python
df_titanic['Sex'] = df_titanic['Sex'].map({'male':1,'female':0})
df_titanic['category_title'] = df_titanic['category_title'].map({'normal_title':1,'unique_title':0})
df_titanic
```

- Label Encoding: digunakan jika terdapat kategori kelas namun bukan merupakan integer, melainkan string (contoh: 1st,2nd,3rd --> yang merupakan kelas 1,2,3)

```python
df_dummy = pd.DataFrame(['1st','1st','2nd','3rd','2nd','1st','3rd'],columns=['Ranking'])
df_dummy
```

- Mean Encoding: mengencode data yang memiliki banyak sekali kategori dan tidak memiliki hirarki. Mean endocing ini **harus** digunakan dengan teknik validasi yang tepat seperti: 1. train-test split, 2. cross-validation. 

**HARUS**, karena kalau pakai teknik encoding ini, seolah kita sudah 'tau' peluang kejadian terhadap variable target, padahal tugas untuk mencari peluang/prediksi terhadap variable target adalah tugas model ML. Then, this is like cheating since the model will learn from a variable that contains the target in itself. Jika kita paksa gunakan, akan menyebabkan over-fitting pada model ML kita nanti. 

But if this is the case we might not even need a ML. Instead, we can use this variable as a single powerful predictor for this target. Reference: [Link](https://towardsdatascience.com/dealing-with-categorical-variables-by-using-target-encoder-a0f1733a4c69)

#Batch 10: Numerical Data Scaling

Numerical Data Scaling = adalah metode untuk merubah data numerical menjadi sebuah skala yang sama dengan variable lainnya. Jika kita buat scatter plot pada data Gaji dan Umur, maka sumbu grafiknya cenderung akan pendek, hal ini disebebkan karena Umur memiliki rentan yang pendek dibandingkan dengan Gaji, sehingga mereka tidak memiliki skala yang sama.

contoh grafik:

image.png

Dan, inti dari KNN adalah ingin mencari titik dengan karakteristik yang **paling mirip**, maka dibutuhkan variable yang memiliki skala yang sama. 

Perlu diingat, metode ini tidak berlaku untuk merubah data scaling pada variable target (y), melainkan untuk variable independent (X).

Why this is important?:
Meskipun perbedaan skala data tidak mempengaruhi (seperti 1M = 100CM, maka 2 hal tersebut adalah hal yang sama), namun jika tidak discaling, maka sense of distence KNN akan terganggu, sehingga hasil dan grafik akan tidak proporsional. Sehingga jika kita scaling data tersebut, KKN akan menilai variabel-variabel yang sudah discaling menjadi lebih objektif terhadap hasil prediksi pada variable targetnya.

How to Scaling:
1. Minmax Scaling = merubah semua angka pada suatu kolom menjadi rentan 0-1.
Rumus:

image.png
2. Standar Scaling = mengubah semua angka dalam column, sehingga distrubsinya memiliki rataan 0 dan standar deviasi 1. 
Rumus:

image.png

```python
#Import library data scaling
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
#Membuat variable X dan Y kembali
X = df_titanic[['Pclass','Age','category_title','C','S','Q','SibSp','Parch','Fare']]
y = df_titanic['Survived']

#Import scaler
scaler = StandardScaler()

#Fit scaler ke X
scaler.fit(X)
X_scaled = scaler.transform(X)

#Import model
model = KNeighborsClassifier()

#Model Fitting
model.fit(X_scaled,y)

#Model Predict
pred = model.predict(X_scaled)

#Model Summary dengan melihat kesamaan hasil Predict ML vs Y
(pred==y).mean()
```
