```python
#Bikin rumus jarak
def distance_formula (x1,x2,y1,y2):
  x_kuadrat = pow((x2 - x1),2)
  y_kuadrat = pow((y2 - y1),2)
  d = pow((x_kuadrat + x_kuadrat), 1/2)
  return (d)

distance_formula(1,3,2,4)
```

```python
# Importing
import sklearn
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the dataset
bos_hou = load_boston()
 
# Create the dataframe
column_name = bos_hou.feature_names
df_boston = pd.DataFrame(bos_hou.data)
df_boston.columns = column_name
df_boston.head()
```

```python
#How to check outlier with boxplot
import seaborn as sns
sns.boxplot(df_boston['DIS'])
```

```python
#Position of the Outlier
import numpy as np
print(np.where(df_boston['DIS']>10))
```

```python
#How to check outlier with Scatter plot
fig, ax = plt.subplots(figsize = (18,10))
ax.scatter(df_boston['INDUS'], df_boston['TAX'])
 
# x-axis label
ax.set_xlabel('(Proportion non-retail business acres)/(town)')
 
# y-axis label
ax.set_ylabel('(Full-value property-tax rate)/( $10,000)')
plt.show()
```

```python
#Position of the Outlier
print(np.where((df_boston['INDUS']>20) & (df_boston['TAX']>600)))
```

```python
df_boston['DIS'].describe()
```

```python
Q1 = np.percentile(df_boston['DIS'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df_boston['DIS'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
```

image.png

```python
print(Q1)
print(Q3)
print(IQR)
```

```python
upper_limit= Q3+1.5*IQR
lower_limit= Q1-1.5*IQR
```

```python
print(upper_limit)
print(lower_limit)
```

```python
# Upper bound
upper = np.where(df_boston['DIS'] >= upper_limit)
# Lower bound
lower = np.where(df_boston['DIS'] <= lower_limit)
```

```python
print(upper)
print(lower)
```

```python
#remove the outlier
df_boston.drop(upper[0], inplace = True)
df_boston.drop(lower[0], inplace = True)
```

```python
outlier = np.where(df_boston['DIS']>10)
df_boston.drop(outlier[0], inplace = True)
sns.boxplot(df_boston['DIS'])
```
